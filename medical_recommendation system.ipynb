## Train Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

### devide dataset into input(x) and output(y) column

```python
x=data.drop("prognosis",axis=1)
y=data["prognosis"]
```

### Encoding output column using LabelEncoder

```python
le=LabelEncoder()
le.fit(y)
Y=le.transform(y)
```

### devide x and Y into training and tetsting dataset

```python
x_train,x_test,y_train,y_test=train_test_split(x,Y,test_size=0.3,random_state=0)
```

```python
x_train
```

```python
Y
```

## Training top model

```python
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
#create a dictionary to store model
models={
    'svc':SVC(kernel='linear'),
    "Randomforest":RandomForestClassifier(n_estimators=100,random_state=42),
    "GradientBoosting":GradientBoostingClassifier(n_estimators=100,random_state=42),
    "Kneighbors":KNeighborsClassifier(n_neighbors=5),
    "Multinomial":MultinomialNB()
}

for model_name,model in models.items():
    #train model
    model.fit(x_train,y_train)
    
    #test model
    prediction=model.predict(x_test)
    
    #calculate accuracy score
    accuracy=accuracy_score(y_test,prediction)
    
    #calculate confusion matrix
    cm=confusion_matrix(y_test,prediction)
    
    print(f"{model_name} accuracy:{accuracy}")
    print(f"{model_name} confusion matrix:")
    print(np.array2string(cm,separator=","))
```

## Single prediction

```python
svc=SVC(kernel='linear')
svc.fit(x_train,y_train)
ypred=svc.predict(x_test)
accuracy_score(y_test,ypred)
```

```python
def helper(dis):
    desc=description[description['Disease']==dis]['Description']
    desc=" ".join([w for w in desc])
    
    pre=precaution[precaution['Disease']==dis][['Precaution_1','Precaution_2','Precaution_3','Precaution_4']]
    pre=[col for col in pre.values]
    
    med=medications[medications['Disease']==dis]['Medication']
    med=[med for med in med.values]
    
    die=diets[diets['Disease']==dis]['Diet']
    die=[die for die in die.values]
    
    wrkout=workout[workout['disease']==dis]['workout']
    
    return desc,pre,med,die,wrkout
```
